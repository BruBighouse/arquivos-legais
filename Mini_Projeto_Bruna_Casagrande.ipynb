```python
# Importando o arquivo com os dados

df = pd.read_csv('/content/kc_house_data.csv')
```

```python
# Formatando a data (date time)

df['date'] = pd.to_datetime(
    arg = df['date'],
    format = '%Y%m%dT%H%M%S'
)
```

```python
df
```

```python
# Criando coluna de cidade pelo zipcode

zcdb = ZipCodeDatabase()
city = []

for zipcode in df['zipcode']:
  try:
    zc = zcdb[zipcode]
    city.append(zc.city)

  except KeyError:
      if zipcode ==98077:
        city.append(str('Woodinville'))
      else:
        city.append(str('none'))

df['city'] = city        
```

```python
df
```

```python
# Verificando valores nulos

df.isna().sum()
```

```python
# Limpando linhas com quartos e banheiros nulos
df[df['bedrooms'].notna()]
df[df['bathrooms'].notna()]
```

```python
# Removendo duplicatas do ID

df[df['id'].notna()]
```

```python
# Distribuição do preço
px.box(df, x='price')
```

```python
#Criando uma coluna para os padrões de imóvel

df.loc[df['price'] < 321837, 'Standard' ] = 1
df.loc[(df['price'] > 321837) & (df['price'] < 645000), 'Standard' ] = 2
df.loc[(df['price'] > 645000) & (df['price'] < 1127500), 'Standard' ] = 3
df.loc[df['price'] > 1127500, 'Standard' ] = 4

df
```

```python
#Distribuição dos preços em boxplot de padrões
px.box(df, x='Standard',y='price')
```

```python
#Criando a coluna de preço do ft2
df['preco_ft2'] = df['price'] / df['sqft_living']
px.box(df, x='preco_ft2')
```

```python
#Criando parâmetros de preço de ft2
df['padrao_preco_ft2'] = 0
df.loc[df['preco_ft2'] < 182.291667, 'padrao_preco_ft2' ] = 1
df.loc[(df['preco_ft2'] > 182.291667) & (df['preco_ft2'] < 244.644550), 'padrao_preco_ft2' ] = 2
df.loc[(df['preco_ft2'] > 244.644550) & (df['preco_ft2'] < 318.322981), 'padrao_preco_ft2' ] = 3
df.loc[df['preco_ft2'] > 318.322981, 'padrao_preco_ft2' ] = 4
```

```python
# Melhores imóveis
df_melhores = df.loc[(df['padrao_preco_ft2'] <= 2) & (df['Standard']>= 4) & (df['view']>= 4) & (df['bedrooms']>= 3) & (df['bathrooms']>= 2)]
df_melhores
```

```python
df_meelhores = df_melhores.head()
map = df_melhores
map = px.scatter_mapbox(df_melhores, lat='lat', lon='long',size='Standard',color='Standard', hover_name='id')
map.update_layout(mapbox_style ='open-street-map')
map.show()
```
